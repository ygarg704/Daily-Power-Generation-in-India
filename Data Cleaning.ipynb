```python
ds_2 = pd.read_csv('Datasets/file_02.csv')
ds_2.info()
```

```python
ds_2 = ds_2.drop(['index', 'Nuclear Generation Actual (in MU)', 'Nuclear Generation Estimated (in MU)'], axis=1)
ds_2 = ds_2.dropna()
ds_2 = ds_2.reset_index(drop = True)
ds_2.head()
```

```python
ds_2['Date'] = pd.to_datetime(ds_2['Date'])
ds_2['Year'] = ds_2['Date'].dt.year
ds_2['Month'] = ds_2['Date'].dt.month
ds_2['Day'] = ds_2['Date'].dt.day
ds_2 = ds_2.drop('Date', axis=1)
ds_2.head()
```

```python
cols = ds_2.columns.tolist()
cols = cols[:1] + cols[-3:] + cols[1:-3]
ds_2 = ds_2[cols]
ds_2.head()
```

```python
ds_2.to_csv('Datasets/file_02_clean.csv', index = False)
ds_2.info()
```

```python
ds_2 = ds_2.drop(['Thermal Generation Estimated (in MU)', 'Hydro Generation Estimated (in MU)'], axis=1)
ds_2['Thermal Generation Actual (in MU)'] = [float(str(i).replace(",", "")) for i in ds_2['Thermal Generation Actual (in MU)']]
ds_2.to_csv('Datasets/predict_clean.csv', index = False)
ds_2.head()
```
